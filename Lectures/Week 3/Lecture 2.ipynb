## matplotlib.pyplot ##

```python
#It is standard to import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

#Here we create the x and y values. It is important they are the same length.
xs = [i/10 for i in range(101)]
ys = [3*x-2 for x in xs]

#This creates a plot
plt.plot(xs,ys)
#This outputs the plot
plt.show()
#Run this when you are done to free memory
plt.close()
```

## Customizing Plots ##

```python
#Here we create the x and y values for each curve
xs1 = [i/10 for i in range(101)]
xs2 = [i/10 + 1 for i in range(101)]
xs3 = [i/10 - 1 for i in range(101)]
ys1 = [3*x-2 for x in xs1]
ys2 = [x**2/10 - 3 for x in xs2]
ys3 = [abs(x-2) for x in xs3]

#We can put multiple curves on the same plot. We can color them
plt.plot(xs1,ys1,color='orange')
plt.plot(xs2,ys2,color='purple')
plt.plot(xs3,ys3,'#3f3f3f')
#We can give a title to the plot
plt.title("My plot")
plt.show()
plt.close()

#If we want separate plots, we have to call plt.show() for each plot.

#Plot 1
plt.plot(xs1,ys1,color='orange')
plt.title("Plot 1")
plt.show()
plt.close()

#Plot 2
plt.plot(xs2,ys2,color='purple')
plt.title("Plot 2")
plt.show()
plt.close()

#Plot 3
plt.plot(xs3,ys3,'#3f3f3f')
plt.title("Plot 3")
plt.show()
plt.close()
```

## Scatter Plots #

```python
#We can create scatter plots
xs1 = [i/10 for i in range(101)]
xs2 = [i/10 + 1 for i in range(101)]
xs3 = [i/10 - 1 for i in range(101)]
ys1 = [3*x-2 for x in xs1]
ys2 = [x**2/10 - 3 for x in xs2]
ys3 = [abs(x-2) for x in xs3]

#We can put multiple curves on the same plot. We can color them and change their shape
plt.scatter(xs1,ys1,color='orange',marker='o')
plt.scatter(xs2,ys2,color='purple',marker='x')
plt.scatter(xs3,ys3,color='#3f3f3f',marker='s')
#We can give a title to the plot
plt.title("My plot")
plt.show()
plt.close()
```

## Histograms ##

```python
#We can also create histograms
xs = [i//10 for i in range(101)]

#We can create our histogram
plt.hist(xs)
plt.show()
plt.close()
```

We can use many possible collections of objects in the list when making a histogram. These include characters in a string.

I will only show the code for the unit circle. We have to be careful when creating this, as plt will try and connect points by default. This can lead to an unexpected line when connecting the top and bottom halves of the circle. One way to fix this is to call plt.plot() twice.

```python
#Here is the circle
xs = [i/100 for i in range(-100,101)]
yupper = [(1-x**2)**(1/2) for x in xs] 
ylower = [-(1-x**2)**(1/2) for x in xs]

plt.plot(xs,yupper, color='b')
plt.plot(xs,ylower, color='b')
plt.show()
plt.close()
```
