The numpy library is helpful for doing math computations. Many of the standard functions in math are in this library.

```python
import numpy as np

#Square Root
print(np.sqrt(2))

#Trig Functions
print(np.sin(3))
print(np.cos(3))
print(np.tan(3))

#Natural Log
print(np.log(10))
#Common Log
print(np.log10(10))

#Exponentials
print(np.exp(2))
```

## Arrays ##

Numpy arrays function like lists, with a few key differences. First, numpy arrays apply operations elementwise. Consider the computations below for lists vs arrays.

```python
mylist = [1,2,3,4]
myarray = np.array(mylist)

print(2*mylist)
print(2*myarray)

print(mylist+mylist+mylist)
print(myarray + 2*myarray - myarray**3)
```

Arrays also consist of objects of one type. If you put in objects of multiple types, then Python will attempt to convert them to the same type.

```python
print(['steerwer',121213,13.2323,2])
#Notice these are all converted to strings
print(np.array(['steerwer',121213,13.2323,2]))


#You can append elements to lists
mylist.append(7)
print(mylist)
#You cannot append elements to arrays
myarray.append(7)
```

Arrays can also be multi-dimensional. A 2D array functions like a matrix.

```python
mymat = [[1,2,3],[2,3,4],[2,3,4]]
print(mymat)
mymat = np.array(mymat)
print(mymat)

#This gives the element in row 0, column 2
print(mymat[0,2])
#This gives the entire row 1
print(mymat[1,:])
#This gives the entire column 2
print(mymat[:,2])
```

If you want an array of a certain size, there are a few ways to initialize the array.

```python
#All entries are zeros in a 6 x 4 matrix
M = np.zeros((6,4))

#This gives the dimensions of M
n,m = M.shape

#We can populate M with new values in a loop
for i in range(n):
    for j in range(m):
        
        M[i,j] = i+j
print(M)

#Array of all ones
print(np.ones(M.shape))

#You can choose the type for the initialized array as well. 
print(np.ones(M.shape,dtype=bool))

print(np.ones(M.shape,dtype=str))

print(np.ones(M.shape,dtype=int))

print(np.ones(M.shape,dtype=float))
```

Add the $2$ matrices below in numpy.

```python
matrix1 = np.array([[1,2],[3,4]])
matrix2 = np.array([[4,6],[7,8]])

print(matrix1 + matrix2)
```

You can also multiply matrices using numpy.

```python
matrix1 = np.array([[1,2],[3,4]])
matrix2 = np.array([[4,6],[7,8]])
#Numpy defaults to elementwise multiplication
print(matrix1*matrix2)

#Matrix Multiplication is done with the @ symbol
print(matrix1@matrix2)
#Matrix Transpose is done with .T
print(matrix1.T)
```

We can also compute some common linear algebra quantities in numpy.

```python
#Here, we will be careful to set the type as float. This is because numpy will not convert to floats when computing the things below.
matrix1 = np.array([[1,2],[3,4]],dtype=float)
matrix2 = np.array([[4,6],[7,8]],dtype=float)

#Find the inverse

#We do not use this for inverse, since it is elementwise
print((matrix1**-1))

#Instead we find the inverse using this function
inv = np.linalg.inv(matrix1)
print(inv)

#Notice we do not get the identity matrix back exactly. Numpy uses numerical methods and sometimes gives small numbers instead of 0
print(inv@matrix1)

#We can round with this function. If we round to 5 decimal places, we get the identity matrix back
print(np.round(inv@matrix1,5))
```

We can also find the eigenvalues and eigenvectors

```python
matrix1 = np.array([[1,2],[3,4]],dtype=float)
#Find the eigenstuffs
eigenvalues, eigenvectors = np.linalg.eig(matrix2)

#The eigenvalues and corresponding eigenvectors are returned in the same order
print(eigenvalues)
print(eigenvectors)


matrix2 = np.array([[1,1],[0,1]],dtype=float)

#Find eigenstuffs
eigenvalues, eigenvectors = np.linalg.eig(matrix2)

#We have to be careful with defective matrices. Numpy may return the same eigenvector scaled by a constant
print(eigenvalues)
print(eigenvectors)
```

```python
#Numpy's eigenvalue and eigenvector function is quite fast, but sometimes gives near zero values for 0.
mat = np.array([[i+j for i in range(30)] for j in range (30)])

print(np.linalg.eig(mat))
```

```python
#We can also compute te determinant quickly (but sometimes the float precision will give nearly correct values)
matrix1 = np.array([[1,2],[3,4]])
print(np.linalg.det(matrix1))

mat = np.array([[i+j for i in range(30)] for j in range (30)])
print(np.linalg.det(mat))
```

## Exercise 2 ##

Find the line of best fit for the data  below in numpy. (Recall the line of best fit coefficients can be found by the following computation).

$$ \vec{b} = (M^{T}M)^{-1}M^{T}\vec{y} $$

```python
M = np.ones((6,2))

for i in range(6):
    M[i,1] = i

y = np.array([1,3,7,10,11,15])

#These are the coefficients
b = np.linalg.inv(M.T@M)@M.T@y
print(b)

#Plot the data
import matplotlib.pyplot as plt

xs = [0,1,2,3,4,5]
ys = [1,3,7,10,11,15]

plt.scatter(xs,ys)

#These are the predictions from the line of best fit
yhat = [b[0] + b[1]*x for x in xs]

plt.plot(xs,yhat,color='r')
plt.show()
plt.close()
```
